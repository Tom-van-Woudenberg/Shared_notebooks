![figuur 1](data/data_interactive_influence_lines/Structure_book.gif)

```python
import sympy as sp
```

```python
import numpy as np
import matplotlib.pylab as plt
from ipywidgets import widgets, interact
```

```python
EI, x = sp.symbols('EI, x',real=True)
```

```python
Bv, Cv, Cm = sp.symbols('Bv, Cv, Cm',real=True)
a, b, L1, L2 = sp.symbols('a, b, L1, L2',positive=True,real=True)
C1, C2, C3, C4 = sp.symbols('C1, C2, C3, C4')
```

# Find support reaction Av #

```python
q = - Bv * sp.SingularityFunction(x,0,-1) - Cv * sp.SingularityFunction(x,L1,-1) + Cm * sp.SingularityFunction(x,L1,-2) + 1 * sp.SingularityFunction(x,a,-1)
V = -sp.integrate(q,x)+C1
M = sp.integrate(V,x)+C2
kappa = M / EI
phi = sp.integrate(kappa,x)+C3
w = sp.integrate(phi,x)+C4
eq1 = sp.Eq(V.subs(x,-1),0)
eq2 = sp.Eq(M.subs(x,-1),0)
eq3 = sp.Eq(w.subs(x,0),0)
eq4 = sp.Eq(w.subs(x,L1),0)
eq5 = sp.Eq(phi.subs(x,L1),0)
eq6 = sp.Eq(M.subs(x,L1+1),0)
eq7 = sp.Eq(V.subs(x,L1+1),0)
sol = sp.solve([eq1,eq2,eq3,eq4,eq5,eq6,eq7],[C1,C2,C3,C4,Bv, Cv, Cm])
Bv_sol = sol[Bv].rewrite(sp.Piecewise)
Bv_numpy = sp.lambdify([L1,a,EI],Bv_sol)
Cv_sol = sol[Cv].rewrite(sp.Piecewise)
Cv_numpy = sp.lambdify([L1,a,EI],Cv_sol)
Cm_sol = sol[Cm].rewrite(sp.Piecewise)
Cm_numpy = sp.lambdify([L1,a,EI],Cm_sol)
V_sol = V.subs(sol)
V_subs = V_sol.subs([(L1,10),(a,2.5),(EI,5000)])
V_numpy = sp.lambdify([L1,a,EI,x],V_sol.rewrite(sp.Piecewise))
M_sol = M.subs(sol)
M_subs = M_sol.subs([(L1,10),(a,2.5),(EI,5000)])
M_numpy = sp.lambdify([L1,a,EI,x],M_sol.rewrite(sp.Piecewise))
```

```python
display(sp.simplify(Bv_sol.rewrite(sp.Piecewise)).expand().subs(a,x))
x_plot = np.linspace(0,10,1000)
plt.plot([0,10],[0,0],color='black',linewidth=2)
plt.plot(x_plot,Bv_numpy(L1=10,EI=5000,a=x_plot),color="blue")
plt.xlabel('$x_F$')
plt.ylabel('Influence factor support reaction at A')
axs = plt.gca()
axs.grid()
axs.invert_yaxis()
title0 = 'Influence lines for support reaction at A'
axs.set_title(title0);
```

##  Finding influence line for M ##

```python
q = 0
V = -sp.integrate(q,x)+C1
M = sp.integrate(V,x)+C2
kappa = M / EI
phi = sp.integrate(kappa,x)+C3 - sp.SingularityFunction(x,a,0)
w = sp.integrate(phi,x)+C4
eq2 = sp.Eq(M.subs(x,0),0)
eq3 = sp.Eq(w.subs(x,0),0)
eq4 = sp.Eq(w.subs(x,L1),0)
eq5 = sp.Eq(phi.subs(x,L1),0)
sol = sp.solve([eq2,eq3,eq4,eq5],[C1,C2,C3,C4])
w_sol = w.subs(sol)
w_subs = w_sol.subs([(L1,10),(a,2),(EI,5000)])
display(sp.simplify(w_subs.rewrite(sp.Piecewise)))
M_inf_numpy = sp.lambdify([L1,a,EI,x],w_sol.rewrite(sp.Piecewise))

x_plot = np.linspace(0,10,1000)
plt.plot([0,10],[0,0],color='black',linewidth=2)
plt.plot(x_plot,M_inf_numpy(L1=10,a=2.5,EI=5000,x=x_plot),color="blue")
plt.xlabel('$x_F$')
plt.ylabel('Influence factor moment')
axs = plt.gca()
axs.grid()
axs.invert_yaxis()
title0 = 'Influence line for moment at $x_M = 2.5$'
axs.set_title(title0);
```

## Find influence line V ##

```python
q = 0
V = -sp.integrate(q,x)+C1
M = sp.integrate(V,x)+C2
kappa = M / EI
phi = sp.integrate(kappa,x)+C3 
w = sp.integrate(phi,x)+C4 + sp.SingularityFunction(x,a,0)
eq2 = sp.Eq(M.subs(x,0),0)
eq3 = sp.Eq(w.subs(x,0),0)
eq4 = sp.Eq(w.subs(x,L1),0)
eq5 = sp.Eq(phi.subs(x,L1),0)
sol = sp.solve([eq2,eq3,eq4,eq5],[C1,C2,C3,C4])
w_sol = w.subs(sol)
w_subs = w_sol.subs([(L1,10),(a,2),(EI,5000)])
display(sp.simplify(w_subs.rewrite(sp.Piecewise)))
V_inf_numpy = sp.lambdify([L1,a,EI,x],w_sol.rewrite(sp.Piecewise))

x_plot = np.linspace(0,10,1000)
plt.plot([0,10],[0,0],color='black',linewidth=2)
plt.plot(x_plot,V_inf_numpy(L1=10,a=2.5,EI=5000,x=x_plot),color="blue")
plt.xlabel('$x_F$')
plt.ylabel('Influence factor shear force')
axs = plt.gca()
axs.grid()
axs.invert_yaxis()
title0 = 'Influence line for shear force at $x_V = 2.5$'
axs.set_title(title0);
```
